```python
import selenium
import pandas as pd
import time
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.in/')
```

```python
title =driver.find_element(By.XPATH,'//input[@class="nav-input nav-progressive-attribute"]')
title.send_keys("laptop")
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span')
search.click()
```

```python
laptop_title =[]

titles = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
for i in titles:
    title = i.text
    laptop_title.append(title)

print(laptop_title)
```

```python
print(len(laptop_title))
```

```python
total_ratings =[]

ratings = driver.find_elements(By.XPATH,'//span[@class="a-size-base a-color-secondary"]')
for i in ratings:
    rating = i.text
    total_ratings.append(rating)

print(total_ratings)
```

```python
print(len(total_ratings))
```

```python
total_prices =[]

price = driver.find_elements(By.XPATH,'//span[@class="a-price"]')
for i in price:
    prices = i.text
    total_prices.append(prices)

print(total_prices)
```

```python
print(len(total_prices))
```

```python
import selenium
import pandas as pd
import time
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
Top_quotes =[]

quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote:
    quotes = i.text
    Top_quotes.append(quotes)

print(Top_quotes)
```

```python
Top_quotes = []

start =0
end =10
for page in range(start,end):
    quotes = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quotes:
        Top_quotes.append(i.text)
        
next_Button =driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
next_Button.click ()
time.sleep(3)
print(Top_quotes)        
```

```python
print(len(Top_quotes))
```
